```python
import cv2
import matplotlib.pyplot as plt

# Function to display images in Jupyter
def show_image(img, title="Image", cmap_type=None):
    plt.figure(figsize=(8, 6))
    plt.imshow(img, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()

# Path to your image file
image_path = 'face.jpg'  # Update this path to your local image path

# Load the image and convert to RGB
img = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB)
show_image(img, "Original Image")

# Display original dimensions
print(f'Original Image Dimensions: {img.shape[0]} x {img.shape[1]}')
print(f'Image Data Type: {img.dtype}')

# Resize to smaller and larger sizes
resized_small = cv2.resize(img, (img.shape[1] // 2, img.shape[0] // 2))
resized_large = cv2.resize(img, (img.shape[1] * 2, img.shape[0] * 2))
show_image(resized_small, "Resized Small Image")
show_image(resized_large, "Resized Larger Image")

# Convert to grayscale
gray_img = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2GRAY)
show_image(gray_img, "Grayscale Image", cmap_type='gray')
print(f'Grayscale Image Dimensions: {gray_img.shape[0]} x {gray_img.shape[1]}')
print(f'Grayscale Image Data Type: {gray_img.dtype}')

# Resize grayscale image
resized_gray_small = cv2.resize(gray_img, (gray_img.shape[1] // 2, gray_img.shape[0] // 2))
resized_gray_large = cv2.resize(gray_img, (gray_img.shape[1] * 2, gray_img.shape[0] * 2))
show_image(resized_gray_small, "Resized Grayscale Small Image", cmap_type='gray')
show_image(resized_gray_large, "Resized Grayscale Larger Image", cmap_type='gray')

# Threshold to create a binary image
_, binary_img = cv2.threshold(gray_img, 128, 255, cv2.THRESH_BINARY)
show_image(binary_img, "Binary Image", cmap_type='gray')
print(f'Binary Image Data Type: {binary_img.dtype}')

# Resize binary image
resized_binary_small = cv2.resize(binary_img, (binary_img.shape[1] // 2, binary_img.shape[0] // 2))
resized_binary_large = cv2.resize(binary_img, (binary_img.shape[1] * 2, binary_img.shape[0] * 2))
show_image(resized_binary_small, "Resized Binary Small Image", cmap_type='gray')
show_image(resized_binary_large, "Resized Binary Larger Image", cmap_type='gray')
```
