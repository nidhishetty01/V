```python
import cv2
import matplotlib.pyplot as plt

# Function to display images in Jupyter
def show_image(img, title="Image"):
    plt.figure(figsize=(8, 6))
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()

# Load the image and convert it to grayscale
image_path = 'face.jpg'
img = cv2.imread(image_path)
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply threshold to create a binary image
_, binary_img = cv2.threshold(gray_img, 128, 255, cv2.THRESH_BINARY)
colored_img = cv2.cvtColor(binary_img, cv2.COLOR_GRAY2BGR)

# Draw geometric shapes and text
cv2.line(colored_img, (0, 0), (100, 100), (255, 0, 0), 5)
cv2.arrowedLine(colored_img, (0, 100), (100, 0), (0, 255, 0), 5)
cv2.rectangle(colored_img, (0, 100), (100, 200), (0, 0, 255), 5)
cv2.circle(colored_img, (50, 50), 20, (255, 255, 0), 5)
cv2.putText(colored_img, 'Hello World', (100, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

# Display the final image
show_image(colored_img, "Geometric Operations on Image")
```
