```python
# Practical 03: Resizing Images and Checking Data Types
import cv2
import matplotlib.pyplot as plt

# Function to display images and their information
def display_image_info(img, title):
    height, width = img.shape[:2]
    
    # Create a figure with size based on the image dimensions
    plt.figure(figsize=(width / 100, height / 100))
    
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB
    plt.title(title)
    plt.axis('off')
    plt.show()
    
    print(f'{title} - Dimensions: {height} x {width} | Data Type: {img.dtype}')

# Path to your image file
image_path = 'face.jpg'  # Update this path with your image file location

# Load the original image
img = cv2.imread(image_path)
if img is None:
    print("Error: Unable to load image. Please check the file path and format.")
else:
    print('Original Image:')
    display_image_info(img, 'Original Image')

    # Resize the image to small size
    print('Resizing to Small Image:')
    new_width = int(img.shape[1] * 0.5)
    new_height = int(img.shape[0] * 0.5)
    resized_small = cv2.resize(img, (new_width, new_height))
    display_image_info(resized_small, 'Resized Small Image')

    # Resize the image to larger size
    print('Resizing to Larger Image:')
    new_width_large = int(img.shape[1] * 2)
    new_height_large = int(img.shape[0] * 2)
    resized_large = cv2.resize(img, (new_width_large, new_height_large))
    display_image_info(resized_large, 'Resized Large Image')

    # Convert the image to grayscale
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    print('Grayscale Image:')
    display_image_info(gray_img, 'Grayscale Image')

    # Resize the grayscale image to small size
    print('Resizing Grayscale to Small Image:')
    resized_gray_small = cv2.resize(gray_img, (new_width, new_height))
    display_image_info(resized_gray_small, 'Resized Grayscale Small Image')

    # Resize the grayscale image to larger size
    print('Resizing Grayscale to Large Image:')
    resized_gray_large = cv2.resize(gray_img, (new_width_large, new_height_large))
    display_image_info(resized_gray_large, 'Resized Grayscale Large Image')

    # Apply simple thresholding to create a binary image
    _, binary_img = cv2.threshold(gray_img, 128, 255, cv2.THRESH_BINARY)
    print('Binary Image:')
    display_image_info(binary_img, 'Binary Image')

    # Resize the binary image to small size
    print('Resizing Binary to Small Image:')
    resized_binary_small = cv2.resize(binary_img, (new_width, new_height))
    display_image_info(resized_binary_small, 'Resized Binary Small Image')

    # Resize the binary image to larger size
    print('Resizing Binary to Large Image:')
    resized_binary_large = cv2.resize(binary_img, (new_width_large, new_height_large))
    display_image_info(resized_binary_large, 'Resized Binary Large Image')
```
